## çŸ¥è¯†åº“æ„å»º
æ­¤æ¨¡å—è´Ÿè´£è¯»å–æ‚¨çš„ Markdown æ–‡ä»¶ï¼Œå°†å…¶åˆ†å‰²æˆå—ï¼Œä½¿ç”¨ bge-m3 æ¨¡å‹è¿›è¡ŒåµŒå…¥ï¼Œå¹¶æ„å»ºä¸€ä¸ª Chroma å‘é‡æ•°æ®åº“ã€‚

````python
# ç¬¬1æ­¥ï¼šå¯¼å…¥åº“ï¼Œè·å–ç¡…åŸºæµåŠ¨api-keyï¼Œä¸ºè°ƒç”¨åµŒå…¥æ¨¡å‹å‡†å¤‡
import os
from langchain_community.vectorstores import Chroma
from langchain_openai import OpenAIEmbeddings
from langchain_community.document_loaders import TextLoader
from langchain.text_splitter import MarkdownHeaderTextSplitter
# å»ºè®®å°† API Key è®¾ç½®ä¸ºç¯å¢ƒå˜é‡ï¼Œé¿å…ç›´æ¥æš´éœ²åœ¨ä»£ç ä¸­
# ä»ç¯å¢ƒå˜é‡è·å– SILICONFLOW_API_KEY
api_key = os.getenv("SILICONFLOW_API_KEY")
if not api_key:
    raise ValueError("è¯·è®¾ç½® SILICONFLOW_API_KEY ç¯å¢ƒå˜é‡")
````

````python
# ç¬¬2æ­¥ï¼šåŠ è½½å’Œåˆ†å‰²æ–‡æ¡£
# è·å–çŸ¥è¯†æ–‡ä»¶è·¯å¾„
import pathlib
# è·å–å½“å‰Jupyter Notebookçš„å·¥ä½œç›®å½•
# åœ¨æ ‡å‡†Jupyterç¯å¢ƒä¸­ï¼Œ.cwd() é€šå¸¸å°±æ˜¯.ipynbæ–‡ä»¶æ‰€åœ¨çš„ç›®å½•
current_dir = pathlib.Path.cwd()

# ä½¿ç”¨ / æ“ä½œç¬¦æ¥æ„å»ºè·¯å¾„ï¼Œéå¸¸ç›´è§‚å’Œä¼˜é›…
# è¿™ä¼šä»å½“å‰ç›®å½•å‡ºå‘ï¼Œå¯»æ‰¾åä¸º'docs'çš„æ–‡ä»¶å¤¹ï¼Œå†å¯»æ‰¾é‡Œé¢çš„æ–‡ä»¶
file_path = current_dir / "docs" / "skincare_products.md"

# æˆ‘ä»¬å¯ä»¥æ‰“å°ä¸€ä¸‹ï¼Œçœ‹çœ‹å®ƒè§£æå‡ºçš„å®Œæ•´è·¯å¾„æ˜¯ä»€ä¹ˆ
print(f"å½“å‰å·¥ä½œç›®å½•: {current_dir}")
print(f"æ„å»ºçš„æ–‡ä»¶è·¯å¾„: {file_path}")
print(f"æ–‡ä»¶æ˜¯å¦å­˜åœ¨: {file_path.exists()}") # æ£€æŸ¥è·¯å¾„æ˜¯å¦æ­£ç¡®

# åŠ è½½Markdownæ–‡ä»¶
loader = TextLoader(file_path, encoding="utf-8")
docs = loader.load()

# å®šä¹‰Markdownçš„åˆ‡åˆ†è§„åˆ™
headers_to_split_on = [
    ("#", "Header 1"),
    ("##", "Header 2"),
    ("###", "Header 3"),
]

# ä½¿ç”¨Markdownæ ‡é¢˜è¿›è¡Œåˆ†å‰²
markdown_splitter = MarkdownHeaderTextSplitter(headers_to_split_on=headers_to_split_on)
md_header_splits = markdown_splitter.split_text(docs[0].page_content)

print(f"æ–‡æ¡£è¢«æˆåŠŸåˆ†å‰²æˆ {len(md_header_splits)} ä¸ªå—ã€‚")
# æ‰“å°ç¬¬ä¸€ä¸ªå—ä½œä¸ºç¤ºä¾‹
print("\n--- ç¬¬ä¸€ä¸ªæ•°æ®å—ç¤ºä¾‹ ---")
print(md_header_splits[0].page_content)
print("\n--- å®ƒçš„å…ƒæ•°æ® ---")
print(md_header_splits[0].metadata)
````

````python
# ç¬¬3æ­¥:åˆå§‹åŒ–åµŒå…¥æ¨¡å‹å¹¶åˆ›å»ºchromaå‘é‡åº“
# åˆå§‹åŒ–ç¡…åŸºæµåŠ¨çš„ BAAI/bge-large-zh-v1.5 åµŒå…¥æ¨¡å‹
# LangChainçš„OpenAIEmbeddingsç±»å¯ä»¥é€šè¿‡é…ç½®base_urlæ¥å…¼å®¹å¤§éƒ¨åˆ†OpenAI APIæ ‡å‡†çš„æ¨¡å‹æœåŠ¡
import shutil  # Pythonæ ‡å‡†åº“ï¼Œæä¾›äº†è®¸å¤šé«˜çº§çš„æ–‡ä»¶å’Œç›®å½•æ“ä½œåŠŸèƒ½ã€‚

embedding_model = OpenAIEmbeddings(
    model="BAAI/bge-large-zh-v1.5",
    api_key=os.getenv("SILICONFLOW_API_KEY"),
    base_url="https://api.siliconflow.cn/v1"
)

# å®šä¹‰å‘é‡æ•°æ®åº“çš„æŒä¹…åŒ–è·¯å¾„
persist_directory = './chroma_db_skincare'

# æ£€æŸ¥å‘é‡æ•°æ®åº“ç›®å½•æ˜¯å¦å·²ç»å­˜åœ¨
if os.path.exists(persist_directory):
    print(f"æ£€æµ‹åˆ°å·²å­˜åœ¨çš„å‘é‡æ•°æ®åº“ '{persist_directory}'ï¼Œæ­£åœ¨åˆ é™¤...")
    # ä½¿ç”¨ shutil.rmtree æ¥é€’å½’åˆ é™¤ç›®å½•åŠå…¶æ‰€æœ‰å†…å®¹
    shutil.rmtree(persist_directory)
    print("æ—§æ•°æ®åº“åˆ é™¤æˆåŠŸã€‚")
# --------------------

# åˆ›å»ºå¹¶æŒä¹…åŒ–å‘é‡æ•°æ®åº“
# è¿™ä¼šå°†åˆ†å‰²åçš„æ–‡æ¡£è¿›è¡ŒåµŒå…¥å¹¶å­˜å…¥ChromaDB
print("æ­£åœ¨åˆ›å»ºå‘é‡æ•°æ®åº“ï¼Œè¯·ç¨å€™...")
vectordb = Chroma.from_documents(
    documents=md_header_splits,
    embedding=embedding_model,
    persist_directory=persist_directory
)
print(f"å‘é‡æ•°æ®åº“åˆ›å»ºæˆåŠŸå¹¶å·²ä¿å­˜è‡³ '{persist_directory}' æ–‡ä»¶å¤¹ï¼")
````

## çŸ¥è¯†åº“æŸ¥è¯¢
æ­¤æ¨¡å—åŠ è½½å·²åˆ›å»ºçš„å‘é‡åº“ï¼Œå¹¶æ„å»ºä¸€ä¸ªæ£€ç´¢å™¨ï¼ˆRetrieverï¼‰ï¼Œç”¨äºæŸ¥è¯¢ç›¸å…³ä¿¡æ¯

````python
# åŠ è½½å‘é‡åº“å¹¶åˆ›å»ºæ£€ç´¢å™¨
# åŠ è½½æŒä¹…åŒ–çš„å‘é‡æ•°æ®åº“
# æ³¨æ„ï¼šå¿…é¡»ä½¿ç”¨ä¸åˆ›å»ºæ—¶ç›¸åŒçš„ embedding_model
print("æ­£åœ¨ä»æœ¬åœ°åŠ è½½å‘é‡æ•°æ®åº“...")
vectordb = Chroma(
    persist_directory=persist_directory,
    embedding_function=embedding_model
)

# åˆ›å»ºä¸€ä¸ªæ£€ç´¢å™¨ï¼Œå¹¶è®¾ç½®å®ƒåªè¿”å›æœ€ç›¸å…³çš„1ä¸ªç»“æœ
retriever = vectordb.as_retriever(search_kwargs={"k": 1})

print("æ£€ç´¢å™¨å·²å‡†å¤‡å°±ç»ªï¼")

# --- åŠŸèƒ½éªŒè¯æµ‹è¯• ---
query = "å“ªæ¬¾äº§å“é€‚åˆç†¬å¤œå…šç”¨æ¥ä¿®æŠ¤é»‘çœ¼åœˆï¼Ÿ"
search_results = retriever.invoke(query)

print(f"\n--- å¯¹æŸ¥è¯¢ '{query}' çš„æ£€ç´¢ç»“æœ ---")
# ç»“æœæ˜¯ä¸€ä¸ªDocumentå¯¹è±¡çš„åˆ—è¡¨
print(search_results[0].page_content)
print(f"\næ¥æºå…ƒæ•°æ®: {search_results[0].metadata}")
````

## æ„å»ºæœ€ç»ˆçš„æŸ¥è¯¢å‡½æ•°
ç°åœ¨ï¼Œæˆ‘ä»¬å°†ä¸Šè¿°æ£€ç´¢åŠŸèƒ½å°è£…æˆæ‚¨è¦æ±‚çš„ query_product_database å‡½æ•°ã€‚

````python
# å®šä¹‰ query_product_database å‡½æ•°
def query_product_database(product_name: str) -> str:
    """
    é€šè¿‡RAGä»å‘é‡çŸ¥è¯†åº“ä¸­æ£€ç´¢äº§å“ä¿¡æ¯ã€‚
    
    å‚æ•°:
    product_name (str): éœ€è¦æŸ¥è¯¢çš„äº§å“åç§°æˆ–ç›¸å…³æè¿°ã€‚
    
    è¿”å›:
    str: æ ¼å¼åŒ–çš„äº§å“è¯¦ç»†ä¿¡æ¯ï¼Œå¦‚æœæ‰¾ä¸åˆ°åˆ™è¿”å›æç¤ºä¿¡æ¯ã€‚
    """
    print(f"[RAG Tool Call] æ­£åœ¨ä»å‘é‡åº“ä¸­æ£€ç´¢å…³äº '{product_name}' çš„ä¿¡æ¯...")
    
    # ä½¿ç”¨ä¹‹å‰åˆ›å»ºçš„retrieverè¿›è¡ŒæŸ¥è¯¢
    results = retriever.invoke(product_name)
    
    if not results:
        return f"äº§å“æ•°æ®åº“ä¸­æœªæ‰¾åˆ°å…³äº '{product_name}' çš„è¯¦ç»†ä¿¡æ¯ã€‚"
    
    # æå–æœ€ç›¸å…³çš„æ–‡æ¡£å†…å®¹å’Œå…ƒæ•°æ®
    top_result_doc = results[0]
    content = top_result_doc.page_content
    metadata = top_result_doc.metadata
    
    # æŒ‰ç…§æ‚¨å¸Œæœ›çš„æ ¼å¼ï¼ˆä¿ç•™åŸæœ‰æ ¼å¼ï¼‰ï¼Œå°†æ ‡é¢˜å’Œå†…å®¹ç»„åˆèµ·æ¥
    # ä»å…ƒæ•°æ®ä¸­è·å–äº§å“åï¼ˆHeader 3ï¼‰
    product_title = metadata.get("Header 3", product_name)
    
    # ç»„åˆæˆ Markdown æ ¼å¼çš„å­—ç¬¦ä¸²
    formatted_output = f"### {product_title}\n\n{content}"
    
    return formatted_output

# --- æœ€ç»ˆå‡½æ•°åŠŸèƒ½éªŒè¯ ---
product_to_query = "å…‰æ„Ÿç„•ç™½ç²¾å"
product_info = query_product_database(product_to_query)

print("\n--- query_product_database å‡½æ•°è°ƒç”¨ç»“æœ ---")
print(product_info)

print("\n" + "="*50 + "\n")

# æµ‹è¯•å¦ä¸€ä¸ªäº§å“
product_to_query_2 = "æ°¨åŸºé…¸æ´—é¢å¥¶æ€ä¹ˆæ ·" # ä½¿ç”¨æ¨¡ç³ŠæŸ¥è¯¢
product_info_2 = query_product_database(product_to_query_2)
print("--- æ¨¡ç³ŠæŸ¥è¯¢ç»“æœ ---")
print(product_info_2)
````

### 3.2 Tools (å·¥å…·å®šä¹‰)

## 4. å®æˆ˜ï¼šæ„å»ºå°çº¢ä¹¦æ–‡æ¡ˆç”Ÿæˆ Agent

ç°åœ¨ï¼Œæˆ‘ä»¬å°†æŠŠ System Promptã€å·¥å…·å®šä¹‰å’Œæ¨¡æ‹Ÿå·¥å…·å‡½æ•°æ•´åˆèµ·æ¥ï¼Œæ„å»ºå‡ºèƒ½å¤Ÿè‡ªåŠ¨æ‰§è¡Œçš„ DeepSeek Agent å·¥ä½œæµã€‚æ ¸å¿ƒæ˜¯ `generate_rednote` å‡½æ•°ï¼Œå®ƒé€šè¿‡ä¸€ä¸ªå¾ªç¯æ¥æ¨¡æ‹Ÿ Agent çš„ `Thought-Action-Observation` è¿‡ç¨‹ã€‚

````python
import json
import re

def generate_rednote(product_name: str, tone_style: str = "æ´»æ³¼ç”œç¾", max_iterations: int = 5) -> str:
    """
    ä½¿ç”¨ DeepSeek Agent ç”Ÿæˆå°çº¢ä¹¦çˆ†æ¬¾æ–‡æ¡ˆã€‚
    
    Args:
        product_name (str): è¦ç”Ÿæˆæ–‡æ¡ˆçš„äº§å“åç§°ã€‚
        tone_style (str): æ–‡æ¡ˆçš„è¯­æ°”å’Œé£æ ¼ï¼Œå¦‚"æ´»æ³¼ç”œç¾"ã€"çŸ¥æ€§"ã€"ææ€ª"ç­‰ã€‚
        max_iterations (int): Agent æœ€å¤§è¿­ä»£æ¬¡æ•°ï¼Œé˜²æ­¢æ— é™å¾ªç¯ã€‚
        
    Returns:
        str: ç”Ÿæˆçš„çˆ†æ¬¾æ–‡æ¡ˆï¼ˆJSON æ ¼å¼å­—ç¬¦ä¸²ï¼‰ã€‚
    """
    
    print(f"\nğŸš€ å¯åŠ¨å°çº¢ä¹¦æ–‡æ¡ˆç”ŸæˆåŠ©æ‰‹ï¼Œäº§å“ï¼š{product_name}ï¼Œé£æ ¼ï¼š{tone_style}\n")
    
    # å­˜å‚¨å¯¹è¯å†å²ï¼ŒåŒ…æ‹¬ç³»ç»Ÿæç¤ºè¯å’Œç”¨æˆ·è¯·æ±‚
    messages = [
        {"role": "system", "content": SYSTEM_PROMPT},
        {"role": "user", "content": f"è¯·ä¸ºäº§å“ã€Œ{product_name}ã€ç”Ÿæˆä¸€ç¯‡å°çº¢ä¹¦çˆ†æ¬¾æ–‡æ¡ˆã€‚è¦æ±‚ï¼šè¯­æ°”{tone_style}ï¼ŒåŒ…å«æ ‡é¢˜ã€æ­£æ–‡ã€è‡³å°‘5ä¸ªç›¸å…³æ ‡ç­¾å’Œ5ä¸ªè¡¨æƒ…ç¬¦å·ã€‚è¯·ä»¥å®Œæ•´çš„JSONæ ¼å¼è¾“å‡ºï¼Œå¹¶ç¡®ä¿JSONå†…å®¹ç”¨markdownä»£ç å—åŒ…è£¹ï¼ˆä¾‹å¦‚ï¼š```json{{...}}```ï¼‰ã€‚"}
    ]
    
    iteration_count = 0
    final_response = None
    
    while iteration_count < max_iterations:
        iteration_count += 1
        print(f"-- Iteration {iteration_count} --")
        
        try:
            # è°ƒç”¨ DeepSeek APIï¼Œä¼ å…¥å¯¹è¯å†å²å’Œå·¥å…·å®šä¹‰
            response = client.chat.completions.create(
                model="deepseek-chat",
                messages=messages,
                tools=TOOLS_DEFINITION, # å‘ŠçŸ¥æ¨¡å‹å¯ç”¨çš„å·¥å…·
                tool_choice="auto" # å…è®¸æ¨¡å‹è‡ªåŠ¨å†³å®šæ˜¯å¦ä½¿ç”¨å·¥å…·
            )

            response_message = response.choices[0].message
            
            # **ReActæ¨¡å¼ï¼šå¤„ç†å·¥å…·è°ƒç”¨**
            if response_message.tool_calls: # å¦‚æœæ¨¡å‹å†³å®šè°ƒç”¨å·¥å…·
                print("Agent: å†³å®šè°ƒç”¨å·¥å…·...")
                messages.append(response_message) # å°†å·¥å…·è°ƒç”¨ä¿¡æ¯æ·»åŠ åˆ°å¯¹è¯å†å²
                
                tool_outputs = []
                for tool_call in response_message.tool_calls:
                    function_name = tool_call.function.name
                    # ç¡®ä¿å‚æ•°æ˜¯åˆæ³•çš„JSONå­—ç¬¦ä¸²ï¼Œå³ä½¿å·¥å…·ä¸è¦æ±‚å‚æ•°ï¼Œä¹Ÿéœ€è¦ä¼ é€’ç©ºå­—å…¸
                    function_args = json.loads(tool_call.function.arguments) if tool_call.function.arguments else {}

                    print(f"Agent Action: è°ƒç”¨å·¥å…· '{function_name}'ï¼Œå‚æ•°ï¼š{function_args}")
                    
                    # æŸ¥æ‰¾å¹¶æ‰§è¡Œå¯¹åº”çš„æ¨¡æ‹Ÿå·¥å…·å‡½æ•°
                    if function_name in available_tools:
                        tool_function = available_tools[function_name]
                        tool_result = tool_function(**function_args)
                        print(f"Observation: å·¥å…·è¿”å›ç»“æœï¼š{tool_result}")
                        tool_outputs.append({
                            "tool_call_id": tool_call.id,
                            "role": "tool",
                            "content": str(tool_result) # å·¥å…·ç»“æœä½œä¸ºå­—ç¬¦ä¸²è¿”å›
                        })
                    else:
                        error_message = f"é”™è¯¯ï¼šæœªçŸ¥çš„å·¥å…· '{function_name}'"
                        print(error_message)
                        tool_outputs.append({
                            "tool_call_id": tool_call.id,
                            "role": "tool",
                            "content": error_message
                        })
                messages.extend(tool_outputs) # å°†å·¥å…·æ‰§è¡Œç»“æœä½œä¸º Observation æ·»åŠ åˆ°å¯¹è¯å†å²
                
            # **ReAct æ¨¡å¼ï¼šå¤„ç†æœ€ç»ˆå†…å®¹**
            elif response_message.content: # å¦‚æœæ¨¡å‹ç›´æ¥è¿”å›å†…å®¹ï¼ˆé€šå¸¸æ˜¯æœ€ç»ˆç­”æ¡ˆï¼‰
                print(f"[æ¨¡å‹ç”Ÿæˆç»“æœ] {response_message.content}")
                
                # --- START: æ·»åŠ  JSON æå–å’Œè§£æé€»è¾‘ ---
                json_string_match = re.search(r"```json\s*(\{.*\})\s*```", response_message.content, re.DOTALL)
                
                if json_string_match:
                    extracted_json_content = json_string_match.group(1)
                    try:
                        final_response = json.loads(extracted_json_content)
                        print("Agent: ä»»åŠ¡å®Œæˆï¼ŒæˆåŠŸè§£ææœ€ç»ˆJSONæ–‡æ¡ˆã€‚")
                        return json.dumps(final_response, ensure_ascii=False, indent=2)
                    except json.JSONDecodeError as e:
                        print(f"Agent: æå–åˆ°JSONå—ä½†è§£æå¤±è´¥: {e}")
                        print(f"å°è¯•è§£æçš„å­—ç¬¦ä¸²:\n{extracted_json_content}")
                        messages.append(response_message) # è§£æå¤±è´¥ï¼Œç»§ç»­å¯¹è¯
                else:
                    # å¦‚æœæ²¡æœ‰åŒ¹é…åˆ° ```json å—ï¼Œå°è¯•ç›´æ¥è§£ææ•´ä¸ª content
                    try:
                        final_response = json.loads(response_message.content)
                        print("Agent: ä»»åŠ¡å®Œæˆï¼Œç›´æ¥è§£ææœ€ç»ˆJSONæ–‡æ¡ˆã€‚")
                        return json.dumps(final_response, ensure_ascii=False, indent=2)
                    except json.JSONDecodeError:
                        print("Agent: ç”Ÿæˆäº†éJSONæ ¼å¼å†…å®¹æˆ–éMarkdown JSONå—ï¼Œå¯èƒ½è¿˜åœ¨æ€è€ƒæˆ–å‡ºé”™ã€‚")
                        messages.append(response_message) # éJSONæ ¼å¼ï¼Œç»§ç»­å¯¹è¯
                # --- END: æ·»åŠ  JSON æå–å’Œè§£æé€»è¾‘ ---
            else:
                print("Agent: æœªçŸ¥å“åº”ï¼Œå¯èƒ½éœ€è¦æ›´å¤šäº¤äº’ã€‚")
                break
                
        except Exception as e:
            print(f"è°ƒç”¨ DeepSeek API æ—¶å‘ç”Ÿé”™è¯¯: {e}")
            break
    
    print("\nâš ï¸ Agent è¾¾åˆ°æœ€å¤§è¿­ä»£æ¬¡æ•°æˆ–æœªèƒ½ç”Ÿæˆæœ€ç»ˆæ–‡æ¡ˆã€‚è¯·æ£€æŸ¥Promptæˆ–å¢åŠ è¿­ä»£æ¬¡æ•°ã€‚")
    return "æœªèƒ½æˆåŠŸç”Ÿæˆæ–‡æ¡ˆã€‚"
````

### æ ¼å¼åŒ– å°çº¢ä¹¦æ–‡æ¡ˆ

**æ ¼å¼åŒ–å‡½æ•° `format_rednote_for_markdown` çš„åŠŸèƒ½ï¼š**

1. è§£æ JSON å­—ç¬¦ä¸²ã€‚
2. æå–æ ‡é¢˜ã€æ­£æ–‡ã€æ ‡ç­¾å’Œè¡¨æƒ…ç¬¦å·ã€‚
3. å°†å®ƒä»¬ç»„åˆæˆä¸€ä¸ªæ˜“è¯»çš„ Markdown æ ¼å¼çš„æ–‡æœ¬ã€‚


**å·¥ä½œæ–¹å¼ï¼š**

1. **è§£æ JSON**ï¼šä½¿ç”¨ `json.loads()` å°†è¾“å…¥çš„å­—ç¬¦ä¸²è½¬æ¢ä¸º Python å­—å…¸ã€‚å¦‚æœè§£æå¤±è´¥ï¼Œä¼šè¿”å›ä¸€ä¸ªé”™è¯¯ä¿¡æ¯ã€‚
2. **æå–æ•°æ®**ï¼šä½¿ç”¨ `.get()` æ–¹æ³•ä»å­—å…¸ä¸­å®‰å…¨åœ°æå– `title`ã€`body` å’Œ `hashtags`ã€‚ä½¿ç”¨ `.get()` çš„å¥½å¤„æ˜¯ï¼Œå¦‚æœæŸä¸ªé”®ä¸å­˜åœ¨ï¼Œå®ƒä¼šè¿”å›ä¸€ä¸ªé»˜è®¤å€¼ï¼ˆä¾‹å¦‚ `None` æˆ–ç©ºåˆ—è¡¨ï¼‰ï¼Œè€Œä¸æ˜¯æŠ›å‡º `KeyError`ã€‚
3. **æ„å»º Markdown æ ‡é¢˜**ï¼šå°† `title` æ ¼å¼åŒ–ä¸º Markdown çš„äºŒçº§æ ‡é¢˜ (`## Title`)ã€‚
4. **å¤„ç†æ­£æ–‡**ï¼šç›´æ¥ä½¿ç”¨ `body`ã€‚ç”±äºå°çº¢ä¹¦æ­£æ–‡ä¸­çš„æ¢è¡Œå¾ˆé‡è¦ï¼Œæˆ‘ä»¬ä¿ç•™å®ƒä»¬ã€‚
5. **å¤„ç† Hashtags**ï¼šå°† `hashtags` åˆ—è¡¨ä¸­çš„æ¯ä¸ªæ ‡ç­¾ç”¨ç©ºæ ¼è¿æ¥èµ·æ¥ï¼Œå½¢æˆä¸€è¡Œã€‚
6. **è¡¨æƒ…ç¬¦å· (Emojis)**ï¼šåœ¨å°çº¢ä¹¦çš„å®é™…å‘å¸ƒä¸­ï¼Œè¡¨æƒ…ç¬¦å·é€šå¸¸å·²ç»åµŒå…¥åœ¨æ ‡é¢˜å’Œæ­£æ–‡ä¸­äº†ã€‚è¿™ä¸ªå‡½æ•°æ²¡æœ‰å•ç‹¬åˆ—å‡ºå®ƒä»¬ï¼Œå› ä¸ºè¿™é€šå¸¸ä¸æ˜¯æœ€ç»ˆå‘å¸ƒæ ¼å¼çš„ä¸€éƒ¨åˆ†ã€‚å¦‚æœéœ€è¦ï¼Œå¯ä»¥å–æ¶ˆæ³¨é‡Šç›¸å…³ä»£ç æ¥å•ç‹¬æ˜¾ç¤ºå®ƒä»¬ã€‚
7. **è¿”å›ç»“æœ**ï¼šè¿”å›æ‹¼æ¥å¥½çš„ Markdown å­—ç¬¦ä¸²ï¼Œå¹¶ä½¿ç”¨ `.strip()` å»é™¤å¯èƒ½å­˜åœ¨äºæœ«å°¾çš„å¤šä½™ç©ºç™½ã€‚

````python
import json

def format_rednote_for_markdown(json_string: str) -> str:
    """
    å°† JSON æ ¼å¼çš„å°çº¢ä¹¦æ–‡æ¡ˆè½¬æ¢ä¸º Markdown æ ¼å¼ï¼Œä»¥ä¾¿äºé˜…è¯»å’Œå‘å¸ƒã€‚

    Args:
        json_string (str): åŒ…å«å°çº¢ä¹¦æ–‡æ¡ˆçš„ JSON å­—ç¬¦ä¸²ã€‚
                           é¢„è®¡æ ¼å¼ä¸º {"title": "...", "body": "...", "hashtags": [...], "emojis": [...]}

    Returns:
        str: æ ¼å¼åŒ–åçš„ Markdown æ–‡æœ¬ã€‚
    """
    try:
        data = json.loads(json_string)
    except json.JSONDecodeError as e:
        return f"é”™è¯¯ï¼šæ— æ³•è§£æ JSON å­—ç¬¦ä¸² - {e}\nåŸå§‹å­—ç¬¦ä¸²ï¼š\n{json_string}"

    title = data.get("title", "æ— æ ‡é¢˜")
    body = data.get("body", "")
    hashtags = data.get("hashtags", [])
    # è¡¨æƒ…ç¬¦å·é€šå¸¸å·²ç»èå…¥æ ‡é¢˜å’Œæ­£æ–‡ä¸­ï¼Œè¿™é‡Œå¯ä»¥é€‰æ‹©æ˜¯å¦å•ç‹¬åˆ—å‡º
    # emojis = data.get("emojis", []) 

    # æ„å»º Markdown æ–‡æœ¬
    markdown_output = f"## {title}\n\n" # æ ‡é¢˜ä½¿ç”¨äºŒçº§æ ‡é¢˜
    
    # æ­£æ–‡ï¼Œä¿ç•™æ¢è¡Œç¬¦
    markdown_output += f"{body}\n\n"
    
    # Hashtags
    if hashtags:
        hashtag_string = " ".join(hashtags) # å°çº¢ä¹¦æ ‡ç­¾é€šå¸¸æ˜¯ç©ºæ ¼åˆ†éš”
        markdown_output += f"{hashtag_string}\n"
        
    # å¦‚æœéœ€è¦ï¼Œå¯ä»¥å•ç‹¬åˆ—å‡ºè¡¨æƒ…ç¬¦å·ï¼Œä½†é€šå¸¸å®ƒä»¬å·²ç»åŒ…å«åœ¨æ ‡é¢˜å’Œæ­£æ–‡ä¸­
    # if emojis:
    #     emoji_string = " ".join(emojis)
    #     markdown_output += f"\nä½¿ç”¨çš„è¡¨æƒ…ï¼š{emoji_string}\n"
        
    return markdown_output.strip() # å»é™¤æœ«å°¾å¤šä½™çš„ç©ºç™½
````
