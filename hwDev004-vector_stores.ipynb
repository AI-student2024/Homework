#### Feat1:支持多个文档的loader、split、embedding和store

```python
from langchain.document_loaders import TextLoader
from langchain_openai import OpenAIEmbeddings
from langchain.text_splitter import CharacterTextSplitter
from langchain.vectorstores import Chroma

# 定义文档路径列表
document_paths = ['../tests/state_of_the_union.txt', '../tests/the_old_man_and_the_sea.txt']  # 假设有多个文档

# 初始化用于存储所有分割后文档的列表
all_documents = []

# 实例化文本分割器，分割器将每个文档分割成200个字符的块
text_splitter = CharacterTextSplitter(chunk_size=200, chunk_overlap=0)

# 循环处理每个文档路径
for path in document_paths:
    # 加载文档内容
    raw_documents = TextLoader(path).load()
    # 使用文本分割器分割文档，并将结果加入到all_documents列表中
    documents = text_splitter.split_documents(raw_documents)
    # 将分割后的文本加入到总列表中
    all_documents.extend(documents)

# 实例化 OpenAI 嵌入模型
embeddings_model = OpenAIEmbeddings(model="text-embedding-ada-002")

# 使用嵌入模型将所有分割后的文本转换为嵌入向量，并存储在向量数据库Chroma中
db = Chroma.from_documents(all_documents, embeddings_model)
```

```python
# 查询向量数据库：
query = "What did the president say about Ketanji Brown Jackson"
docs = db.similarity_search(query)
len(docs)
print(docs[0].page_content)
```

```python
# 查询向量数据库：
query = "老人在梦里梦见什么了？"
docs = db.similarity_search(query)
print(len(docs))
print(docs[2].page_content)
```

```python
import gradio as gr
from langchain.document_loaders import TextLoader
# 导入其他必要的库，依赖于支持的文档类型
from langchain_openai import OpenAIEmbeddings
from langchain.text_splitter import CharacterTextSplitter
from langchain.vectorstores import Chroma
import os

# 定义一个函数来处理和加载不同格式的文档
def load_document(file_path):
    _, file_extension = os.path.splitext(file_path)
    if file_extension.lower() == '.txt':
        return TextLoader(file_path).load()
    # 这里可以添加其他格式的处理逻辑，例如.docx, .pdf, .jpg 等
    else:
        raise ValueError("Unsupported file type")

# 定义一个函数来处理上传的文件
def process_file(file):
    # 保存上传的文件到本地
    with open("uploaded_file", "wb") as f:
        f.write(file)
    
    # 加载文档
    documents = load_document("uploaded_file")
    
    # 初始化文本分割器
    text_splitter = CharacterTextSplitter(chunk_size=200, chunk_overlap=0)
    
    # 分割文档
    split_documents = text_splitter.split_documents([documents])
    
    # 初始化嵌入模型
    embeddings_model = OpenAIEmbeddings(model="text-embedding-ada-002")
    
    # 获取文档的嵌入向量
    embeddings = [embeddings_model.get_embeddings(doc) for doc in split_documents]
    
    # 初始化向量存储数据库
    db = Chroma()
    
    # 存储嵌入向量
    for embedding in embeddings:
        db.add(embedding)
    
    return "File has been processed and vectors are stored."

# 定义一个函数来执行向量检索
def search_vectors(query):
    # 初始化嵌入模型
    embeddings_model = OpenAIEmbeddings(model="text-embedding-ada-002")
    
    # 获取查询的嵌入向量
    query_embedding = embeddings_model.get_embeddings(query)
    
    # 初始化向量存储数据库
    db = Chroma()
    
    # 执行向量检索
    results = db.search(query_embedding, top_k=3)
    
    return results

# 创建 Gradio 界面
def create_interface():
    with gr.Blocks() as demo:
        gr.Markdown("### Document Upload")
        with gr.Row():
            file_upload = gr.File(file_count="single", label="Upload Document")
            submit_button = gr.Button("Process File")
        
        gr.Markdown("### Vector Search")
        with gr.Row():
            query_input = gr.Textbox(label="Enter query")
            search_button = gr.Button("Search")
        
        output = gr.Textarea(label="Results")
        
        # 将函数绑定到界面组件
        file_upload.change(process_file, inputs=file_upload, outputs=output)
        submit_button.click(process_file, inputs=file_upload, outputs=output)
        query_input.change(search_vectors, inputs=query_input, outputs=output)
        search_button.click(search_vectors, inputs=query_input, outputs=output)
    
    return demo

# 运行 Gradio 界面
if __name__ == "__main__":
    interface = create_interface()
    interface.launch()
```
